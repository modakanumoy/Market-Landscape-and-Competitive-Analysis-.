```python
import matplotlib.pyplot as plt

platforms = ['Coursera', 'edX', 'Udacity', 'Khan Academy', 'LinkedIn Learning']
courses_offered = [7000, 6400, 500, 10000, 16000]

plt.figure(figsize=(10, 6))
plt.barh(platforms, courses_offered, color='skyblue')
plt.xlabel('Number of Courses Offered')
plt.title('Number of Courses Offered by Platform')
plt.show()
```

```python
user_base = [142, 35, 14, 70, 900]

plt.figure(figsize=(10, 6))
plt.barh(platforms, user_base, color='lightgreen')
plt.xlabel('User Base (millions)')
plt.title('User Base by Platform')
plt.show()
```

```python
completion_rate = [55.4, 90, 35, 96, 15]

plt.figure(figsize=(10, 6))
plt.barh(platforms, completion_rate, color='salmon')
plt.xlabel('Completion Rate (%)')
plt.title('Completion Rate by Platform')
plt.show()
```

```python
user_satisfaction = [4.7, 4.9, 4.5, 4.7, 5]

plt.figure(figsize=(10, 6))
plt.barh(platforms, user_satisfaction, color='lightcoral')
plt.xlabel('User Satisfaction')
plt.title('User Satisfaction by Platform')
plt.show()
```
